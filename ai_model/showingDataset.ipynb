```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import random
from matplotlib.image import imread
```

```python
# Define the path to the directory containing the subfolders
data_dir = "images/train"
# Define the subfolders
subfolders = ["CNV", "DME", "DRUSEN", "NORMAL"]
# Define the number of images to select from each subfolder
num_images_per_subfolder = 5

# Initialize a list to store the selected image paths
selected_images = []
```

```python
for subfolder in subfolders:
    # Construct the path to the current subfolder
    subfolder_path = os.path.join(data_dir, subfolder)
    
    # Get a list of all the image filenames in the subfolder
    images = os.listdir(subfolder_path)
    
    # Select a random sample of images from the subfolder
    selected_images.extend([os.path.join(subfolder_path, img) for img in random.sample(images, num_images_per_subfolder)])
```

```python
# Function to convert RGB image to grayscale
def rgb2gray(rgb):
    return np.dot(rgb[..., :3], [0.2989, 0.5870, 0.1140])
```

```python
# Plot the selected images
plt.figure(figsize=(15, 10))
for i, image_path in enumerate(selected_images, 1):
    plt.subplot(4, 5, i)
    img = imread(image_path)
    if len(img.shape) > 2 and img.shape[2] == 3:  # Check if image is RGB
        img = rgb2gray(img)
    plt.imshow(img, cmap='gray')
    plt.title(os.path.basename(os.path.dirname(image_path)))
    plt.axis("off")
    
plt.tight_layout()
plt.show()
```
